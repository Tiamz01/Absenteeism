#### it isimportance to understand why reason for absence is numerical instead word. it is realize to is was emcoded to numerical other, therefore it a categoricl nominal data. to work with it, we have to encode it using dummies to ad a numerical value to it.

```python
reason_column = pd.get_dummies(df['Reason for Absence'], dtype=int)
```

```python
reason_column['check'] = reason_column.sum(axis=1)
reason_column.head()
```

```python
reason_column = pd.get_dummies(df['Reason for Absence'], drop_first = True, dtype='int')
```

```python
reason_column
```

```python
# Dropping the reason clumn from the df to avoid multicollinearity

df = df.drop('Reason for Absence', axis=1)
```

```python
df.head()
```

#### grouping the reason columns to avoid overpopulated dataframe. Group the reason columns to their generlised form.

```python
reason_type_1= reason_column.loc[:, 1:14].max(axis=1)
reason_type_2= reason_column.loc[:, 15:17].max(axis=1)
reason_type_3= reason_column.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_column.loc[:, 22:].max(axis=1)
```

```python
#concatenate the reason type columns to existing df
df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4])
df.columns.values
```

```python
column_names= {0:'reason_1', 1:'reason_2', 2:'reason_3', 3:'reason_3'}

df = df.rename(columns=column_names, inplace=True)
```
